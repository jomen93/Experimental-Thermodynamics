## Primera parte

Para la primera parte lo que se hace es ver la temperatura de ebullición del agua de $200 ml$ de Agua, y luego se le agregan cantidades de $0.5 g$ de sal y se evidencia si la temperatura de ebullición cambia , se tomaron do series de datos, entonces se tiene que 


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
sal = np.arange(0.,5.5,0.5)
T1 = np.array([95.,94.8,94.9,94.9,95.,95.1,95.2,95.3,95.4,95.5,95.7])
T2 = np.array([95.,94.9,95.,95.1,95.2,95.3,95.3,95.4,95.5,95.6,95.6])
m = 200 # ml 
T = (T1+T2)/2.

yErr = 0.1/2; tc1Err = 0.5/2
plt.plot(sal,T,"b.",label = "promedio")
plt.errorbar(sal, T, yErr, tc1Err,ecolor = "b",fmt='+b',capsize = 2)
plt.legend()
plt.grid(True)
plt.xlabel("Sal [g]")
plt.ylabel("Temperatura $[^{o}C]$")
plt.title("Cambio de sal, medicion temperatura")

plt.savefig("CTemp")
```
