```python
def Error_linealizacion(x,y):
    yp = np.mean(y)*np.ones(len(y))
    ei2 = np.sum((y-yp)**2)
    ey = np.sqrt(ei2/(len(y)-2))
    xi2 = np.sum(x**2); Xi2 = np.sum(x)**2
    db = ey*np.sqrt((len(y))/(len(y)*xi2-Xi2)) #pendiente
    da = ey*np.sqrt((xi2)/(len(y)*xi2-Xi2)) #punto de corte
    return np.array([da,db])

```

$$T(t)-T_{f} =(T_{i}-T_{f})e^{-t/\tau}$$

Ahora calculamos el rea de cada cilindro para comparar los tiempo de enfriamiento y su rea superficial

con la siguiente frmula 

$$A_{s} = 2\pi r(2h +r)$$

y el error con 

$$\Delta A_{s} = |A_{s}|\sqrt{\left(\frac{\Delta r}{r}\right)^{2}+\left(\frac{\Delta h}{h}\right)^{2}}$$

```python
def AreaS(r,h):
    return 2*np.pi*r*(h + r)/100. 

def AreaSErr(r,h):
    dh = dr = 0.5
    return np.abs(AreaS(r,h))*np.sqrt((dr/r)**2 +(dh/h)**2)/100.

```

Ahora el calculo de la conductancia trmica unitaria 

a partir de la siguiente frmula

$$h = \frac{mc}{\tau A_{s}}$$

y el error

$$\Delta h = |h|\sqrt{\left(\frac{\Delta \tau}{\tau}\right)^{2}+\left(\frac{\Delta A_{s}}{A_{s}}\right)^{2}+\left(\frac{\Delta m}{m}\right)^{2}}$$


```python
def h(m,tau,A):
    c = 123.91 #J/gC
    return (m*c)/(tau*A)

def hErr(dtau,tau,dA,A,m):
    dm = 0.05
    return np.sqrt((dtau/tau)**2 +(dA/A)**2 +(dm/m)**2)
    
m1 = 70.2
m2 = 34.9
m3 = 28.6
m4 =14.9

hc1 = h(m1,tauc1,Ac1) # W/cm2
hc2 = h(m2,tauc1,Ac1) # W/cm2
hc3 = h(m3,tauc1,Ac1) # W/cm2
hc4 = h(m4,tauc1,Ac1) # W/cm2

print "h1 =", hc1, "Error", hErr(tauc1Err,tauc1,dAc1,Ac1,m1)
print "h2 =", hc2, "Error", hErr(tauc2Err,tauc2,dAc2,Ac2,m2)
print "h3 =", hc3, "Error", hErr(tauc3Err,tauc3,dAc3,Ac3,m3)
print "h4 =", hc4, "Error", hErr(tauc4Err,tauc4,dAc4,Ac4,m4)
```
